```python
# Limpieza
def limpieza(df):
  print('\n----- Descripción del Dataset -----\n')

  # Primeras 5 filas
  print("Primeras 5 filas del dataset:")
  print(tabulate(df.head(), headers='keys', tablefmt='psql'))
  print("\n")

  # Cantidad de NaNs
  print("Cantidad de NaNs por columna:")
  cant_nans = df.isna().sum().reset_index()
  cant_nans.columns = ['Columna', 'Cantidad de NaNs']
  print(tabulate(cant_nans, headers='keys', tablefmt='psql'))
  print("\n")

  # Verificación de duplicados
  print("Cantidad de filas duplicadas en el DataFrame:")
  duplicados = df.duplicated().sum()
  print(f">>> {duplicados} fila(s) duplicada(s) encontrada(s).\n")

  # Opcional: Mostrar las filas duplicadas (si existen)
  if duplicados > 0:
      print("Filas duplicadas:")
      print(df[df.duplicated()])

  # Info general
  print("Información general del dataset:")
  print(tabulate(df.info(), headers='keys', tablefmt='psql'))
  print("\n")

  # Separar variables numéricas y categóricas
  numericas = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
  categoricas = df.select_dtypes(include=['object']).columns.tolist()

  print("\nVariables numéricas:", numericas)
  print("Variables categóricas:", categoricas)
```

```python
limpieza(tienda)
limpieza(tienda2)
limpieza(tienda3)
limpieza(tienda4)
```

#1. Análisis de facturación



```python
tiendas = {
    'Tienda 1': tienda,
    'Tienda 2': tienda2,
    'Tienda 3': tienda3,
    'Tienda 4': tienda4
}

for nombre, df in tiendas.items():
  ingreso = df['Precio'].sum()
  costo_envio = df['Costo de envío'].sum()
  ganancias = ingreso - costo_envio

  print(f'\n{nombre}')
  print(f'Ingresos totales: ${ingreso:,.2f}')
  print(f'Costos de envio: ${costo_envio:,.2f}')
  print(f'Ganancia neta: ${ganancias:,.2f}')
```

# 4. Productos más y menos vendidos

En este paso, debes identificar qué productos fueron los más vendidos y los menos vendidos en cada tienda. Visualiza los resultados para que quede claro qué productos destacaron en ventas en cada tienda.

```python
def producto_mas_vendido(df, nombre):
    """
    Imprime y grafica el producto más y menos vendido de una tienda.

    Parámetros
    ----------
    df : pandas.DataFrame
        Datos de la tienda, debe tener una columna 'Producto'.
    nombre : str
        Nombre de la tienda que aparecerá en los mensajes y el título del gráfico.
    """
    # 1. Conteo de ventas por producto
    conteo = df['Producto'].value_counts()              # Serie: índice=producto, valores=cantidad

    # 2. Máximos y mínimos
    prod_max  = conteo.idxmax()                         # nombre del más vendido
    ventas_max = conteo.max()                           # cantidad del más vendido
    prod_min  = conteo.idxmin()                         # nombre del menos vendido
    ventas_min = conteo.min()                           # cantidad del menos vendido

    # 3. Salida de texto
    print(f'\nTienda {nombre}:')
    print(f'  🥇 Producto más vendido:  {prod_max} ({ventas_max} ventas)')
    print(f'  🥄 Producto menos vendido: {prod_min} ({ventas_min} ventas)')

    # 4. Preparar datos para graficar
    datos_plot = conteo.reset_index()                   # DataFrame con dos columnas
    datos_plot.columns = ['Producto', 'Cantidad']
    datos_plot = datos_plot.sort_values('Cantidad', ascending=False)

    # 5. Gráfico de barras
    plt.figure(figsize=(10, 6))
    sns.barplot(
        data=datos_plot,
        y='Producto',           # Productos en el eje Y para que el texto sea legible
        x='Cantidad',
        palette='viridis'
    )
    plt.title(f'Productos más vendidos - {nombre}')
    plt.xlabel('Cantidad de ventas')
    plt.ylabel('Producto')
    plt.tight_layout()
    plt.show()
```

```python
producto_mas_vendido(tienda, 1)
producto_mas_vendido(tienda2, 2)
producto_mas_vendido(tienda3, 3)
producto_mas_vendido(tienda4, 4)
```

# Graficos

```python
plt.rcParams["figure.figsize"] = (8,5)

urls = {
    "Tienda 1": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv",
    "Tienda 2": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv",
    "Tienda 3": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv",
    "Tienda 4": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv",
}

tiendas = {k: pd.read_csv(v) for k, v in urls.items()}

# KPI por tienda --------------------------------------------------------------
kpi = []
for nombre, df in tiendas.items():
    ingreso  = df["Precio"].sum()
    envio    = df["Costo de envío"].sum()
    beneficio = ingreso - envio
    rating   = df["Calificación"].mean()
    envio_med = df["Costo de envío"].mean()
    kpi.append([nombre, ingreso, beneficio, rating, envio_med])

kpi_df = (pd.DataFrame(kpi, columns=["Tienda","Ingresos","Beneficio",
                                     "Rating_prom","Envio_promio"])
          .set_index("Tienda")
          .sort_values("Beneficio", ascending=False))
kpi_df

kpi_df["Beneficio"].plot.barh()
plt.title("Beneficio neto por tienda")
plt.xlabel("COP")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()

# --- 2.2 Línea: evolución anual de ingresos (si te interesa)
# Convertimos fechas a datetime y sumamos por año
ingresos_anuales = {}
for n, df in tiendas.items():
    df["Fecha de Compra"] = pd.to_datetime(df["Fecha de Compra"], dayfirst=True)
    ingresos_anuales[n] = (df.groupby(df["Fecha de Compra"].dt.year)["Precio"]
                             .sum())
pd.DataFrame(ingresos_anuales).plot(marker="o")
plt.title("Ingresos anuales por tienda")
plt.ylabel("COP")
plt.xlabel("Año")
plt.tight_layout()
plt.show()

# --- 2.3 Dispersión: rating vs. coste medio de envío (tamaño = beneficio)
plt.scatter(kpi_df["Envio_promio"], kpi_df["Rating_prom"],
            s=kpi_df["Beneficio"]/2e7,  alpha=0.7)
for i,row in kpi_df.iterrows():
    plt.text(row["Envio_promio"]+200, row["Rating_prom"]+0.01, i)
plt.title("Costo de envío vs satisfacción")
plt.xlabel("Costo de envío medio (COP)")
plt.ylabel("Calificación media")
plt.tight_layout()
plt.show()

fig, ax = plt.subplots(figsize=(6,6))
for n, df in tiendas.items():
    ax.scatter(df["lon"], df["lat"], label=n, alpha=0.2, s=10)
ax.set_title("Dispersión geográfica de ventas")
ax.set_xlabel("Longitud")
ax.set_ylabel("Latitud")
ax.legend()
plt.show()

# (Bonus) Heatmap de densidad con hist2d de Matplotlib
todo = pd.concat(tiendas.values(), ignore_index=True)
plt.hist2d(todo["lon"], todo["lat"], bins=60)
plt.colorbar(label="N° de ventas")
plt.title("Mapa de calor: concentración de ventas")
plt.xlabel("Longitud"); plt.ylabel("Latitud")
plt.tight_layout(); plt.show()



```
